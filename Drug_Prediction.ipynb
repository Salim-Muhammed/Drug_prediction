```python
df1=df.drop(['Sex'],axis=1)
```

```python
df1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1["BP"]=le.fit_transform(df1['BP'])
df1['Cholesterol']=le.fit_transform(df1['Cholesterol'])
df1['Na_to_K']=le.fit_transform(df1['Na_to_K'])
df1
```

```python
x=df1.iloc[:,:-1].values
y=df1.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)

```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
matr=confusion_matrix(y_pred,y_test)
matr
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
labels=['drugX','drugA','drugC','drugY','drugB']
cmd=ConfusionMatrixDisplay(matr,display_labels=labels)
cmd.plot()
```

```python
score=accuracy_score(y_pred,y_test)
score
```

```python
report=classification_report(y_test,y_pred)
print(report)
```
